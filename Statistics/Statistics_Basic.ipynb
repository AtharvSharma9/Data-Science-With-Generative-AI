# Statistics Basic Solution

## Theory Questions

**1. What is statistics, and why is it important?**


Statistics is a branch of mathematics that deals with collecting, organizing, analyzing, interpreting, and presenting data. It helps us understand data patterns and make informed decisions.

Importance:

* Makes sense of complex data.

* Used in fields like business, healthcare, economics, sports, and research.

* Supports evidence-based decisions.

**2. What are the two main types of statistics?**


1. Descriptive Statistics: Summarizes or describes the features of a dataset.

2. Inferential Statistics: Makes predictions or inferences about a population based on a sample of data.

**3. What are descriptive statistics?**


Descriptive statistics are methods for summarizing data using:

* Measures of central tendency (mean, median, mode)

* Measures of dispersion (range, variance, standard deviation)

* Tables, graphs, charts for visualization

* They do not involve making predictions or generalizations beyond the data.

**4. What is inferential statistics?**


Inferential statistics involves drawing conclusions or making predictions about a population based on a sample. It uses:

* Hypothesis testing

* Confidence intervals

* Regression analysis

Example: Predicting election results based on a sample survey.

**5. What is sampling in statistics?**


Sampling is the process of selecting a subset (sample) from a larger population to analyze and make inferences.

Reason: It is often impractical or impossible to study the entire population.

**6. What are the different types of sampling methods?**


* Random Sampling: Every element has an equal chance (e.g., lottery draw).

* Systematic Sampling: Every nth element is selected.

* Stratified Sampling: Population divided into strata; samples taken from each.

* Cluster Sampling: Population divided into clusters, then randomly select clusters.

* Convenience Sampling: Sample chosen based on ease of access.

* Quota Sampling: Sample reflects certain characteristics proportionally.

**7. What is the difference between random and non-random sampling?**


* Random Sampling: Every individual has an equal chance. Reduces bias.

* Non-Random Sampling: Selection based on convenience or judgment. May have bias.

**8. Define and give examples of qualitative and quantitative data.**


* Qualitative Data: Categorical data (non-numeric).

Example: Colors, names, gender, blood type.

* Quantitative Data: Numerical data.

Example: Height, age, income, temperature.

**9. What are the different types of data in statistics?**


1. Qualitative (Categorical)

    * Nominal

    * Ordinal

2. Quantitative (Numerical)

    * Discrete

    * Continuous

**10. Explain nominal, ordinal, interval, and ratio levels of measurement.**


* Nominal: Categories without order (e.g., gender, color)

* Ordinal: Categories with order, but no fixed interval (e.g., ranking: 1st, 2nd)

* Interval: Ordered, equal intervals, no true zero (e.g., temperature in Â°C)

* Ratio: Ordered, equal intervals, true zero exists (e.g., weight, height)

**11. What is the measure of central tendency?**


It represents a single value that describes the center of a dataset:

* Mean (average)

* Median (middle value)

* Mode (most frequent value)

**12. Define mean, median, and mode.**


* Mean: Sum of all values Ã· total number

* Mode: Most frequently occurring value

Helps to:

** 15. What is standard deviation, and why is it important?**


Standard Deviation is the square root of variance. It shows how much the data deviates from the mean.

* Importance: Helps understand data spread or consistency.

**16. Define and explain the term range in statistics.**


Range = Maximum value â€“ Minimum value

It shows the spread of data but is sensitive to outliers.

**17. What is the difference between variance and standard deviation?**


* Variance: Average of squared differences from the mean (in square units).

* Standard Deviation: Square root of variance (same unit as data). Easier to interpret.

**18. What is skewness in a dataset?**


Skewness measures the asymmetry of the data distribution.

* Symmetrical data: Skewness = 0

* Skewed data: Either positively or negatively skewed

**19. What does it mean if a dataset is positively or negatively skewed?**


* Positively Skewed: Tail on the right; mean > median > mode

* Negatively Skewed: Tail on the left; mean < median < mode

**20. Define and explain kurtosis.**


Kurtosis measures the "tailedness" or peakedness of a distribution.

* High kurtosis: Heavy tails (outliers likely)

* Low kurtosis: Light tails (few outliers)

**21. What is the purpose of covariance?**


Covariance shows how two variables change together:

* Positive: Both variables increase together

* Negative: One increases while the other decreases

It helps understand relationships between variables but lacks standard scale.

**22. What does correlation measure in statistics?**


Correlation measures the strength and direction of a linear relationship between two variables, scaled between -1 and 1.

* +1: Perfect positive

* -1: Perfect negative

* 0: No correlation


**24. What are some real-world applications of statistics?**


* Healthcare: Disease prediction, clinical trials

* Business: Market analysis, forecasting, customer behavior

* Sports: Player performance analysis

* Government: Census, policy planning

* Technology: Machine learning models, data science

**Q1 How do you calculate the mean, median, and mode of a dataset?**

**Q4 Implement sampling techniques like random sampling and stratified sampling?**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Create a simple dataset
df = pd.DataFrame({
    'Gender': ['Male', 'Female'] * 5,
    'Age': [20, 22, 25, 30, 28, 27, 31, 35, 29, 33]
})

# Random Sampling
random_sample = df.sample(n=4)
print("Random Sample:\n", random_sample)

# Stratified Sampling
strat_sample, _ = train_test_split(df, stratify=df['Gender'], test_size=0.4)
print("Stratified Sample:\n", strat_sample)
```

**Q8 Generate a dataset and demonstrate positive and negative skewness?**

```python
import matplotlib.pyplot as plt
import numpy as np

# Positive skew
pos_skew = np.random.exponential(scale=2.0, size=1000)
# Negative skew (invert exponential)
neg_skew = -np.random.exponential(scale=2.0, size=1000)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(pos_skew, bins=30, color='orange')
plt.title("Positive Skew")

plt.subplot(1, 2, 2)
plt.hist(neg_skew, bins=30, color='green')
plt.title("Negative Skew")

plt.show()
```

**Q12  Implement and compare simple random sampling and systematic sampling**

**Q13. Mean, median, and mode of grouped data**

```python
import pandas as pd
import statistics

data = {
    "Class Interval": ["0-10", "10-20", "20-30", "30-40"],
    "Frequency": [5, 8, 15, 7]
}

df = pd.DataFrame(data)

# Mid-points
df["Midpoint"] = df["Class Interval"].apply(lambda x: (int(x.split("-")[0]) + int(x.split("-")[1])) / 2)

# Mean
mean = sum(df["Midpoint"] * df["Frequency"]) / sum(df["Frequency"])
print("Grouped Mean:", mean)

# Median and Mode for grouped data would need cumulative frequencies and formulas
```

**Q24 Pearson and Spearman correlation coefficients**

```python
from scipy.stats import pearsonr, spearmanr

x = np.random.normal(50, 5, 100)
y = x + np.random.normal(0, 2, 100)

pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
