**Q1: What is Anomaly Detection?**

Anomaly Detection is the process of identifying patterns in data that do not conform to the expected or normal behavior. These unusual patterns are called anomalies or outliers.
It is widely used in fraud detection, network security, fault detection in systems, and medical diagnosis.

**Types of Anomalies**

**Point Anomalies**

A single data point is significantly different from the rest of the data.

Example:

In credit card transactions, if most users spend between ₹500–₹5,000 per day, but one user suddenly makes a ₹5,00,000 purchase, that transaction is a point anomaly.

In a dataset of student exam scores (mostly between 40–90), a score of 0 or 100 might be a point anomaly.

**Contextual Anomalies**

A data point is anomalous only in a specific context (depends on time, location, or situation).

Example:

A temperature of 35°C is normal in summer but anomalous in winter.

A spike in website traffic during the day may be normal, but the same spike at 3 AM could be an anomaly.

**Collective Anomalies**

A group of related data points is anomalous, even if individual points look normal.

Example:

In a server log, repeated failed login attempts (like 100 attempts in 5 minutes) indicate a potential cyber-attack—though each single login attempt isn’t abnormal.

In stock market data, a sudden unusual sequence of rising and falling prices could represent market manipulation.

**Q2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)**

1. **Isolation Forest (iForest)**

**Approach:**

Based on the idea that anomalies are easier to isolate than normal points.

Uses random decision trees to recursively split data.

Points requiring fewer splits are considered anomalies.

**Suitable Use Cases:**

High-dimensional datasets.

Large-scale problems where efficiency is important (e.g., fraud detection, intrusion detection).

Works well when anomalies are globally different from normal data.

2. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

**Approach:**

A density-based clustering algorithm.

Defines clusters as areas of high point density, and points in low-density regions are treated as anomalies (noise).

**Suitable Use Cases:**

Data with clusters of varying shapes (not just circular).

Spatial data, geographical data, or when anomalies are far from dense regions.

Example: Identifying unusual GPS movement patterns.

3. **Local Outlier Factor (LOF)**

**Approach:**

Density-based local method.

Compares the local density of a point with its neighbors.

If the point’s density is much lower than neighbors, it is an anomaly.

**Suitable Use Cases:**

Detecting contextual anomalies where local density matters.

Data with varying densities (where global methods like Isolation Forest may fail).

Example: Fraudulent transactions within a specific group of customers.

**Comparison Table**

| Method            | Approach                    | Strengths                               | Limitations                 | Best Use Cases                           |
|-------------------|-----------------------------|-----------------------------------------|-----------------------------|------------------------------------------|
| Isolation Forest  | Random splits to isolate points | Fast, scalable, good for high-dim data | May miss local anomalies    | Fraud detection, cybersecurity           |
| DBSCAN            | Density-based clustering      | Finds arbitrary-shaped clusters         | Struggles with high-dim data | GPS data, geographical clustering        |
| LOF               | Local density comparison      | Good for local anomalies                | Computationally expensive   | Fraud within small communities           |

**Q3: Key Components of a Time Series**

A time series is a sequence of data points collected at regular time intervals. It can be broken into four main components:

1. **Trend (T)**

Definition: The long-term upward or downward movement in data over a long period.

Example: The steady increase in the number of internet users worldwide over the past two decades shows a positive trend.

2. **Seasonality (S)**

Definition: Repeating patterns or cycles in data that occur at fixed intervals (daily, monthly, yearly).

Example: Ice cream sales peak every summer and drop during winter.

3. **Cyclical Component (C)**

Definition: Long-term fluctuations in data that are not of fixed length, often related to economic or business cycles.

Example: Stock market booms and recessions occur in cycles but don’t follow a fixed calendar.

4. **Irregular/Random Component (I)**

Definition: Unpredictable, random variations caused by unexpected events.

Example: A sudden drop in tourism due to a natural disaster or pandemic.

**Summary Table**
| Component   | Meaning                        | Example                         |
|-------------|--------------------------------|---------------------------------|
| Trend       | Long-term movement             | Rise in global internet users   |
| Seasonality | Regular repeating pattern      | Ice cream sales peak in summer  |
| Cyclical    | Long-term cycles (not fixed)   | Stock market booms & crashes    |
| Irregular   | Random, unpredictable changes  | Tourism drop due to pandemic    |

**Q4: Stationarity in Time Series**

**Definition:**

A time series is stationary if its statistical properties (mean, variance, autocorrelation) remain constant over time.

*   No trend
*   No seasonality
*   Constant mean & variance

Stationarity is important because many time series models (like ARIMA) assume the series is stationary.

**How to Test for Stationarity**

*   **Visual Inspection**
    *   Plot the series $\rightarrow$ check if mean and variance change over time.

*   **Statistical Tests**
    *   **Augmented Dickey-Fuller (ADF) Test**
        *   Null Hypothesis ($H_0$): Series is non-stationary.
        *   If p-value < 0.05, reject $H_0$ $\rightarrow$ series is stationary.
    *   **KPSS Test (Kwiatkowski–Phillips–Schmidt–Shin)**
        *   Null Hypothesis ($H_0$): Series is stationary.
        *   If p-value < 0.05, reject $H_0$ $\rightarrow$ series is non-stationary.

**How to Transform Non-Stationary Series into Stationary**

*   **Differencing**
    *   Subtract the current value from the previous value.
    *   Example: $Y_t' = Y_t - Y_{t-1}$.
    *   Removes trend.
*   **Detrending**
    *   Fit and remove a trend (linear regression or moving average).
*   **Seasonal Adjustment**
    *   Remove seasonal effects using seasonal differencing or decomposition.
*   **Transformation**
    *   Apply log, square root, or Box-Cox transformation to stabilize variance.
    *   Example: $\log(Y_t)$ reduces heteroscedasticity.

**Q5: Difference between AR, MA, ARIMA, SARIMA, and SARIMAX Models**

1.  **AR (Autoregressive Model)**

    **Structure:**

    Predicts current value using past values of the series.

    **Equation:**

    $Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \epsilon_t$

    **Application:**

    Good for series with correlation between current and past values.

    Example: Stock prices depending on previous day’s prices.

2.  **MA (Moving Average Model)**

    **Structure:**

    Predicts current value using past error terms (shocks/noise).

    **Equation:**

    $Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \epsilon_t$

    **Application:**

    Good for series where random shocks influence future values.

    Example: Modeling unexpected changes in demand.

3.  **ARIMA (Autoregressive Integrated Moving Average)**

    **Structure:**

    Combination of AR + I (differencing) + MA.

    ARIMA(p, d, q)

    *   p = lag order (AR part)
    *   d = differencing order (to make series stationary)
    *   q = order of MA

    **Application:**

    Handles non-stationary time series with trend.

    Example: Forecasting sales data with upward trend.

4.  **SARIMA (Seasonal ARIMA)**

    **Structure:**

    Extends ARIMA to include seasonality.

    SARIMA(p, d, q)(P, D, Q, s)

    *   (p,d,q) $\rightarrow$ non-seasonal part
    *   (P,D,Q) $\rightarrow$ seasonal part
    *   s $\rightarrow$ season length

    **Application:**

    Good for seasonal data like weather, monthly sales, energy consumption.

    Example: Ice cream sales peaking every summer.

5.  **SARIMAX (Seasonal ARIMA with Exogenous Variables)**

    **Structure:**

    SARIMA + external factors (exogenous regressors).

    Equation includes additional predictors (X variables).

    **Application:**

    Used when external variables influence time series.

    Example:

    *   Predicting electricity demand using SARIMAX with temperature as an external variable.
    *   Forecasting sales including marketing spend or holidays.

**Q6: Time Series Decomposition with AirPassengers Dataset**

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the AirPassengers dataset (available in statsmodels or via CSV)
# Dataset contains monthly airline passengers from 1949–1960
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Plot the original series
plt.figure(figsize=(10, 4))
plt.plot(data, label="AirPassengers")
plt.title("Original AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Decompose the time series into Trend, Seasonality, and Residual
decomposition = seasonal_decompose(data['Passengers'], model='additive', period=12)

# Plot decomposition
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

**Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.(Include your Python code and output in the code box below.)**

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv" # Using AirPassengers as a temporary placeholder
data = pd.read_csv(url)

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, sep=';') # Wine Quality data is semicolon-separated


X = data[['alcohol', 'density']].dropna()

# Apply Isolation Forest
# Adjusting contamination if needed based on expected anomaly percentage
iso = IsolationForest(contamination=0.02, random_state=42)
X['anomaly'] = iso.fit_predict(X)

# Mark anomalies (-1 = anomaly, 1 = normal)
anomalies = X[X['anomaly'] == -1]
normal = X[X['anomaly'] == 1]

# Plot anomalies vs normal points
plt.figure(figsize=(10, 6))
plt.scatter(normal['alcohol'], normal['density'],
            c='blue', s=10, label="Normal")
plt.scatter(anomalies['alcohol'], anomalies['density'],
            c='red', s=20, label="Anomaly")
plt.title("Isolation Forest - Wine Quality Anomaly Detection")
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.legend()
plt.show()
```

**Question 8: Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.(Include your Python code and output in the code box below.)**

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Train-Test Split (train until 1959, forecast 1960)
train = data.iloc[:-12]
test = data.iloc[-12:]

model = SARIMAX(train['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))
sarima_fit = model.fit(disp=False)

# Forecast next 12 months
forecast = sarima_fit.forecast(steps=12)

# Plot results
plt.figure(figsize=(10,6))
plt.plot(train.index, train['Passengers'], label="Training Data")
plt.plot(test.index, test['Passengers'], label="Actual Data", color="green")
plt.plot(forecast.index, forecast, label="SARIMA Forecast", color="red")
plt.title("SARIMA Forecast - AirPassengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

**Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib. (Include your Python code and output in the code box below.)**

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris(as_frame=True)
data = iris.data[['sepal length (cm)', 'sepal width (cm)']]

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
labels = lof.fit_predict(data)

# Separate normal points and anomalies
normal = data[labels == 1]
anomalies = data[labels == -1]

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(normal['sepal length (cm)'], normal['sepal width (cm)'],
            c='blue', s=30, label="Normal")
plt.scatter(anomalies['sepal length (cm)'], anomalies['sepal width (cm)'],
            c='red', s=50, label="Anomaly")
plt.title("Local Outlier Factor (LOF) - Iris Dataset")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()
```

**Question 10: You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:

● How would you detect anomalies in this streaming data (Isolation Forest / LOF /DBSCAN)?

● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

● How would you validate and monitor the performance over time?

● How would this solution help business decisions or operations?**

1. Anomaly Detection in Streaming Data

Since data comes every 15 minutes, anomalies should be detected in real-time.

Choice of method:

Isolation Forest → scalable, fast for streaming-like data.

LOF → better for local/contextual anomalies, but slower.

DBSCAN → not ideal for real-time since it needs full batch clustering.

Isolation Forest is best for real-time anomaly detection.

2. Short-Term Forecasting Model

Energy demand is seasonal (daily & weekly patterns).

Models:

ARIMA → only handles trend (not seasonality).

SARIMA → handles seasonality but not external features.

SARIMAX → handles seasonality + external variables (like temperature, region).

SARIMAX is best (since weather & region affect demand).

3. Validation & Monitoring

Split data into train/test (rolling or sliding window).

Metrics: RMSE, MAE, MAPE for forecasts.

For anomalies: Precision, Recall, F1 against known anomalies.

Continuous monitoring → retrain weekly/monthly as new data arrives.

4. Business Impact

Detecting abnormal spikes/drops → prevents grid overloads or blackouts.

Forecasting demand → better resource allocation, energy pricing, and load balancing.

Real-time dashboard → operators can act before problems escalate.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -------------------------
# 1. Simulate Sample Streaming Data
# -------------------------
np.random.seed(42)
date_rng = pd.date_range(start="2023-01-01", periods=200, freq="15T")
energy = 200 + 10*np.sin(np.linspace(0,50,200)) + np.random.normal(0,5,200)

# Add anomaly spikes
energy[50] = 400
energy[120] = 50

df = pd.DataFrame({"timestamp": date_rng, "energy_usage": energy})
df.set_index("timestamp", inplace=True)

# -------------------------
# 2. Anomaly Detection (Isolation Forest)
# -------------------------
iso = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy_usage"]])

# Separate anomalies
anomalies = df[df["anomaly"] == -1]

plt.figure(figsize=(10,5))
plt.plot(df.index, df["energy_usage"], label="Energy Usage")
plt.scatter(anomalies.index, anomalies["energy_usage"], color="red", label="Anomalies")
plt.title("Energy Usage with Anomalies (Isolation Forest)")
plt.xlabel("Time")
plt.ylabel("Energy Demand")
plt.legend()
plt.show()

# -------------------------
# 3. Forecasting with SARIMAX
# -------------------------
train = df.iloc[:-20]
test = df.iloc[-20:]

model = SARIMAX(train["energy_usage"], order=(1,1,1), seasonal_order=(1,1,1,96))  # 96 points per day
sarimax_fit = model.fit(disp=False)

forecast = sarimax_fit.forecast(steps=20)

# -------------------------
# 4. Evaluate Forecast
# -------------------------
mae = mean_absolute_error(test["energy_usage"], forecast)
rmse = np.sqrt(mean_squared_error(test["energy_usage"], forecast))

print("MAE:", mae)
print("RMSE:", rmse)

# Plot forecast
plt.figure(figsize=(10,5))
plt.plot(train.index, train["energy_usage"], label="Training Data")
plt.plot(test.index, test["energy_usage"], label="Actual Data", color="green")
plt.plot(test.index, forecast, label="SARIMAX Forecast", color="red")
plt.title("SARIMAX Forecast - Energy Demand")
plt.xlabel("Time")
plt.ylabel("Energy Demand")
plt.legend()
plt.show()
```
