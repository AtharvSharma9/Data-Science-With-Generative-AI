**Question 1:** What is a Convolutional Neural Network (CNN), and how does it differ from
traditional fully connected neural networks in terms of architecture and performance on
image data?

A Convolutional Neural Network (CNN) is a specialized deep learning model designed to process and analyze grid-like data, especially images. It uses convolution operations to automatically learn spatial features such as edges, textures, shapes, and objects.

How CNNs differ from Fully Connected Neural Networks (FNNs)
1. Architecture Differences

| Aspect | CNN | Fully Connected Neural Network (FNN) |
|---|---|---|
| Connections | Uses local connections (filters/kernels) | Every neuron connected to every neuron (dense) |
| Layers | Convolutional layers, pooling layers, feature maps | Only dense (linear) layers |
| Parameter Count | Much fewer parameters (shared weights) | Very high number of parameters |
| Spatial Information | Preserves spatial structure of images | Flattens image ‚Üí loses spatial relationships |
| Weight Sharing | Yes ‚Üí One filter slides across entire image | No weight sharing |

2. Performance Differences on Image Data

| Factor | CNN | FNN |
|---|---|---|
| Image Feature Extraction | Automatically learns features like edges, shapes | Does not learn spatial features well |
| Computation | Faster and more efficient due to fewer parameters | Slow and computationally expensive |
| Overfitting | Less prone to overfitting | Highly prone to overfitting on images |
| Accuracy | High accuracy in image classification tasks | Lower accuracy compared to CNNs |


**Question 2:** Discuss the architecture of LeNet-5 and explain how it laid the foundation
for modern deep learning models in computer vision. Include references to its original
research paper.

LeNet-5 is a pioneering Convolutional Neural Network (CNN) proposed by Yann LeCun et al. in 1998 in the research paper titled
‚ÄúGradient-Based Learning Applied to Document Recognition.‚Äù

It was designed for handwritten digit recognition (MNIST dataset) and is considered one of the earliest successful deep learning architectures.

**Architecture of LeNet-5**

LeNet-5 consists of 7 layers (excluding input) that alternate between convolution, subsampling, and fully connected layers. The input image size is 32 √ó 32.

Layer-wise Architecture

1. Input Layer

  * 32 √ó 32 grayscale image

2. C1 ‚Äì Convolution Layer

  * 6 feature maps

  * 5 √ó 5 filters

  * Output: 28 √ó 28 √ó 6

  * Extracts local features (edges, curves)

3. S2 ‚Äì Subsampling (Average Pooling)

  * 2 √ó 2 pooling with stride 2

  * Output: 14 √ó 14 √ó 6

  * Reduces spatial size & introduces translation invariance

4. C3 ‚Äì Convolution Layer

  * 16 feature maps

  * 5 √ó 5 filters

  * Output: 10 √ó 10 √ó 16

  * Complex feature extraction

  * Not fully connected to all maps in S2 (controlled connectivity)

5. S4 ‚Äì Subsampling (Average Pooling)

  * Output: 5 √ó 5 √ó 16

6. C5 ‚Äì Fully Connected Convolution Layer

  * 120 feature maps

  * 5 √ó 5 filter covers the entire S4 map

  * Essentially a fully connected layer

7. F6 ‚Äì Fully Connected Layer

  * 84 neurons

  * Uses RBF-like activation (in original paper)

8. Output Layer

  * 10 neurons (digits 0‚Äì9)

**How LeNet-5 Laid the Foundation for Modern Deep Learning Models**

1. Introduced Core CNN Concepts

LeNet-5 introduced:

  * Convolution layers

  * Weight sharing

  * Pooling layers

  * Hierarchical feature extraction

These form the basis of modern CNNs like AlexNet, VGG, ResNet, and Inception.

2. Demonstrated End-to-End Learning

LeNet-5 showed that a network could:

  * Learn features automatically

  * Replace handcrafted features (like SIFT, HOG)

  * Perform classification end-to-end

This idea is the foundation of modern deep learning pipelines.

3. Efficient Training & Use of Gradient Descent

The paper showed the effectiveness of:

  * Backpropagation

  * Stochastic gradient descent

  * Parameter sharing

These are still core methods in all modern deep learning models.

4. Proved CNNs Work for Real-World Applications

LeNet-5 successfully automated handwriting recognition used by:

  * US Banks

  * Postal services

This commercial success inspired later large-scale models like AlexNet (2012), which reignited interest in deep learning.

5. Foundation for Modern Architectures

Modern CNNs expand on LeNet-5‚Äôs concepts:

| LeNet-5 Feature | Modern Extensions |
|---|---|
| Small kernels | Deeper, stacked layers (VGG) |
| Pooling | Max pooling, adaptive pooling |
| Convolution | Residual connections (ResNet) |
| Parameter sharing | Depthwise separable convs (MobileNet) |
| Simple architecture | Complex multi-branch models (Inception) |

**Question 3: **Compare and contrast AlexNet and VGGNet in terms of design principles, number of parameters, and performance. Highlight key innovations and limitations of each.

**1. Introduction**

AlexNet (2012) and VGGNet (2014) are two landmark CNN architectures in computer vision. Both significantly improved performance on the ImageNet Large-Scale Visual Recognition Challenge (ILSVRC), but they differ in design philosophy, depth, and computational requirements.

**2. Comparison Between AlexNet and VGGNet**

A. Design Principles

AlexNet

  * Introduced in 2012 by Krizhevsky et al.

  * 8 layers (5 convolutional + 3 fully connected)

  * Large convolution kernels (11√ó11, 5√ó5)

  * Extensive use of ReLU, which accelerated training

  * Overlapping max-pooling layers

  * Used dropout to reduce overfitting

  * Trained using two GPUs due to hardware limitations

VGGNet

  * Introduced in 2014 by Simonyan & Zisserman

  * Very deep architecture: 16 or 19 layers (VGG16, VGG19)

  * Uses small 3√ó3 filters stacked multiple times

  * Uniform, simple architecture

  * No large kernels; relies on depth for feature extraction

  * Max-pooling used after each block

  * Fully connected layers similar to AlexNet

B. Number of Parameters

| Model | Parameters | Remarks |
|---|---|---|
| AlexNet | ~60 million | Heavy due to fully connected layers |
| VGG16 | ~138 million | More than double AlexNet |
| VGG19 | ~144 million | Extremely large and memory-intensive |

C. Performance

| Model | Top-5 Accuracy on ImageNet | Improvement |
|---|---|---|
| AlexNet (2012) | 84.7% | First deep CNN to outperform traditional methods |
| VGGNet (2014) | 92.7% | Significant improvement due to depth |

3. Key Innovations

**AlexNet Innovations**

1. ReLU Activation

  * Made deep networks trainable and faster than tanh/sigmoid.

2. Dropout Regularization

  * Prevented overfitting significantly.

3. GPU Training

  * First large-scale deep network to use GPUs effectively.

4. Data Augmentation

  * Pioneered extensive augmentation (cropping, flipping).

5. Overlapping Max Pooling

  * Improved feature extraction over standard pooling.

**VGGNet Innovations**

1. Small (3√ó3) Convolutions

  * Three 3√ó3 convs = one 7√ó7 receptive field but with fewer parameters and more non-linearity.

2. Deep Architecture

  * 16‚Äì19 layers showed that depth improves performance.

3. Modular Design

  * Simple and uniform architecture ‚Üí easy to implement and extend.

VGG‚Äôs design inspired many later models:

* ResNet (skip connections)

* Inception

* MobileNet

4. Limitations.

AlexNet Limitations

* Large filter sizes ‚Üí more parameters

* Relatively shallow compared to modern networks

* Used outdated features like Local Response Normalization (LRN)

* Requires two GPUs due to memory limits

VGGNet Limitations

* Very large model size (138M parameters)

* Memory-intensive ‚Üí slow to train

* High computational cost ‚Üí not suitable for mobile/edge devices

* Fully connected layers consume most parameters

**Question 4: **What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data

Transfer learning is a technique in deep learning where a model trained on a large dataset (such as ImageNet) is reused or fine-tuned for a different but related image classification task. Instead of training a CNN from scratch, we start with a pre-trained model that already knows useful features like edges, textures, shapes, and objects.

How Transfer Learning Works

A pre-trained CNN (e.g., VGG, ResNet, MobileNet) is used in two ways:

1. Feature Extraction

* The convolutional layers of the pre-trained model are kept frozen.

* Only the final fully connected layers are replaced and trained.

* Used when dataset is very small.

2. Fine-Tuning

* Some convolutional layers are unfrozen and retrained.

* The model adapts to new domain-specific features.

* Used when dataset is moderate in size.

**How Transfer Learning Reduces Computational Costs**

1. Avoids Training from Scratch

Training large CNNs like ResNet or VGG requires:

* Millions of parameters

* High compute power

* Days or weeks of training

Transfer learning bypasses this because the pre-trained weights are reused.

2. Reuses Learned Low-Level Features

Early layers in CNNs detect:

* Edges

* Gradients

* Corners

* Textures

These features are universal across all images ‚Üí no need to learn them again.

3. Requires Less Training Time

Only the top layers (new classifier) are trained ‚Üí drastically reduces:

* Training time

* GPU usage

* Memory consumption

Training becomes possible even on a single GPU or a normal laptop.

How Transfer Learning Improves Performance with Limited Data

1. Helps Avoid Overfitting

When data is limited, training a deep CNN from scratch leads to overfitting.
A pre-trained model already has generalized filters, so it learns better with few samples.

2. High Accuracy Even With Small Datasets

Models pre-trained on ImageNet (1.2M images) have strong general-purpose knowledge.
This boosts accuracy on small datasets (e.g., medical images, plant disease datasets).

3. Learns Domain-Specific Features Faster

Fine-tuning adapts the model to specific tasks:

* Cancer detection

* Face recognition

* Vehicle classification

This significantly increases accuracy compared to random initialization.

4. Better Feature Representation

The representations learned from massive datasets transfer well across domains, enabling:

* Stronger generalization

* More robust classification

**Question 5:** Describe the role of residual connections in ResNet architecture. How do
they address the vanishing gradient problem in deep CNNs?

Residual connections (also called skip connections) are the core idea behind the ResNet (Residual Network) architecture introduced by He et al. in 2015. They allow the input of a layer to be added directly to the output of a deeper layer, bypassing one or more convolution layers.

This is known as a residual block, represented as:

Output=F(x)+x

Where:

x = input

F(x) = output of stacked convolutional layers

x + F(x) = residual learning

Role of Residual Connections in ResNet
1. Enable Very Deep Networks

Before ResNet, training CNNs deeper than ~20‚Äì30 layers was difficult.
ResNet can reach 50, 101, 152, or even 1000+ layers because of skip connections.

How Residual Connections Solve the Vanishing Gradient Problem
1. Provide a Direct Path for Gradients

In deep CNNs, gradients become extremely small during backpropagation, causing vanishing gradients.

Residual connections create a shortcut so the gradient can flow directly from deeper layers to earlier layers without shrinking.

**Question 6:** Implement the LeNet-5 architectures using Tensorflow or PyTorch to
classify the MNIST dataset. Report the accuracy and training time.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import time

# ---------- LeNet-5 Architecture ----------
class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, kernel_size=5)
        self.pool = nn.AvgPool2d(2, stride=2)
        self.conv2 = nn.Conv2d(6, 16, kernel_size=5)
        # Corrected input size for fc1: 16 * 5 * 5 = 400
        self.fc1 = nn.Linear(16*5*5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = self.pool(x)
        x = torch.relu(self.conv2(x))
        x = self.pool(x)
        # Flatten the tensor. x's shape before this is (N, 16, 5, 5)
        x = x.view(-1, 16*5*5) # Corrected to 16*5*5
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# ---------- Data Loading ----------
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor()
])

trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
testset  = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=64, shuffle=True)
testloader  = DataLoader(testset, batch_size=64, shuffle=False)

# ---------- Training ----------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LeNet5().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

start = time.time()
for epoch in range(5):
    model.train()
    running_loss = 0
    for batch_idx, (images, labels) in enumerate(trainloader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)

        # --- Debugging: Print shapes before loss calculation ---
        if batch_idx == 0 and epoch == 0: # Print only for the first batch of the first epoch
            print(f"Debug: Epoch {epoch+1}, Batch {batch_idx+1}")
            print(f"Debug: outputs shape: {outputs.shape}")
            print(f"Debug: labels shape: {labels.shape}")
        # --------------------------------------------------------

        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {running_loss/len(trainloader):.4f}")
training_time = time.time() - start

# ---------- Testing ----------
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
print(f"Training Time: {training_time:.2f} seconds")
```

dataset (e.g., flowers or animals). Replace the top layers and fine-tune the model.
Include your code and result discussion

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Download and extract the flower dataset
!wget -q https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz
!tar xzf flower_photos.tgz
!mv flower_photos flowers

train_dir = "/content/flowers/train"
test_dir  = "/content/flowers/test"

# Create train and test directories, and move some images for testing
import os
import shutil
import random

if not os.path.exists(train_dir):
    os.makedirs(train_dir)
if not os.path.exists(test_dir):
    os.makedirs(test_dir)

# Assuming 'flowers' directory is now at /content/flowers
source_dir = "/content/flowers"
classes = [d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d)) and d not in ['train', 'test']]

for cls in classes:
    class_path = os.path.join(source_dir, cls)
    images = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
    random.shuffle(images)

    num_train = int(len(images) * 0.8) # 80% for training

    train_class_dir = os.path.join(train_dir, cls)
    test_class_dir = os.path.join(test_dir, cls)

    if not os.path.exists(train_class_dir):
        os.makedirs(train_class_dir)
    if not os.path.exists(test_class_dir):
        os.makedirs(test_class_dir)

    for img_name in images[:num_train]:
        shutil.move(os.path.join(class_path, img_name), os.path.join(train_class_dir, img_name))
    for img_name in images[num_train:]:
        shutil.move(os.path.join(class_path, img_name), os.path.join(test_class_dir, img_name))


train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze base model layers for Feature Extraction
for layer in base_model.layers:
    layer.trainable = False

x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(train_gen.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=5,
    verbose=1
)


for layer in base_model.layers[-4:]:
    layer.trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history_ft = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=5,
    verbose=1
)
loss, acc = model.evaluate(test_gen)
print(f"\nFinal Test Accuracy: {acc*100:.2f}%")

plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title("Accuracy Curve")
plt.show()
```

**Question 9:** Train a GoogLeNet (Inception v1) or its variant using a standard dataset
like CIFAR-10. Plot the training and validation accuracy over epochs and analyze
overfitting or underfitting.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load CIFAR-10 (Resized to 64√ó64)
# -----------------------------
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor()
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset  = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=32, shuffle=True)
testloader  = DataLoader(testset, batch_size=32, shuffle=False)

# -----------------------------
# 2. Load GoogLeNet (Small Variant)
# -----------------------------
model = torchvision.models.googlenet(aux_logits=False, num_classes=10)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_acc = []
val_acc = []

# -----------------------------
# 3. Training Loop
# -----------------------------
epochs = 5
for epoch in range(epochs):
    model.train()
    correct = 0
    total = 0

    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    train_acc.append(correct/total)

    # Validation
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = outputs.max(1)

            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    val_acc.append(correct/total)

    print(f"Epoch {epoch+1} ‚Äî Train Acc: {train_acc[-1]:.3f}, Val Acc: {val_acc[-1]:.3f}")

# -----------------------------
# 4. Plot Accuracy
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(train_acc, label="Training Accuracy")
plt.plot(val_acc, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("GoogLeNet on CIFAR-10")
plt.legend()
plt.show()
```

```python
# ---------------------------------------------------------
#  END-TO-END MEDICAL X-RAY CLASSIFICATION PIPELINE
#  (ResNet50 Transfer Learning + Fine-tuning + Grad-CAM)
# ---------------------------------------------------------

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os


# ---------------------------------------------------------
# Helper function to create dummy dataset for demonstration
# ---------------------------------------------------------
def create_dummy_dataset(base_dir="/content/dataset", img_size=224):
    if os.path.exists(base_dir):
        import shutil
        shutil.rmtree(base_dir)
        print(f"Removed existing {base_dir}")

    classes = ["COVID", "NORMAL", "PNEUMONIA"]
    sets = ["train", "test"]

    for s in sets:
        for cls in classes:
            os.makedirs(os.path.join(base_dir, s, cls), exist_ok=True)

    # Create dummy images
    dummy_image = np.zeros((img_size, img_size, 3), dtype=np.uint8)
    cv2.imwrite(os.path.join(base_dir, "train", "COVID", "sample_0.png"), dummy_image)
    cv2.imwrite(os.path.join(base_dir, "train", "NORMAL", "sample_0.png"), dummy_image)
    cv2.imwrite(os.path.join(base_dir, "train", "PNEUMONIA", "sample_0.png"), dummy_image)

    cv2.imwrite(os.path.join(base_dir, "test", "COVID", "sample.png"), dummy_image)
    cv2.imwrite(os.path.join(base_dir, "test", "NORMAL", "sample.png"), dummy_image)
    cv2.imwrite(os.path.join(base_dir, "test", "PNEUMONIA", "sample.png"), dummy_image)

    print(f"Dummy dataset created at {base_dir}")

# Call the function to create the dummy dataset
create_dummy_dataset()


# ---------------------------------------------------------
# 1. LOAD DATASET (Train/Test)
# ---------------------------------------------------------

train_path = "/content/dataset/train"
test_path  = "/content/dataset/test"

IMG_SIZE = 224
BATCH = 32

train_datagen = ImageDataGenerator(
    rescale=1/255.0,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=False,
)

test_datagen = ImageDataGenerator(rescale=1/255.0)

train_gen = train_datagen.flow_from_directory(
    train_path,
    target_size=(IMG_SIZE, IMG_SIZE),
    class_mode="categorical",
    batch_size=BATCH,
    shuffle=True
)

test_gen = test_datagen.flow_from_directory(
    test_path,
    target_size=(IMG_SIZE, IMG_SIZE),
    class_mode="categorical",
    batch_size=BATCH,
    shuffle=False
)

NUM_CLASSES = len(train_gen.class_indices)
print("Classes:", train_gen.class_indices)


# ---------------------------------------------------------
# 2. LOAD PRETRAINED RESNET-50 (BASE MODEL)
# ---------------------------------------------------------

base_model = ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

# freeze lower layers
for layer in base_model.layers[:140]:
    layer.trainable = False


# ---------------------------------------------------------
# 3. ADD CUSTOM CLASSIFIER HEAD
# ---------------------------------------------------------

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)
output = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=Adam(1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()


# ---------------------------------------------------------
# 4. TRAIN THE MODEL (Feature Extraction + Fine-tuning)
# ---------------------------------------------------------

history = model.fit(
    train_gen,
    epochs=5,
    validation_data=test_gen
)


# ---------------------------------------------------------
# 5. FINE-TUNE DEEPER LAYERS
# ---------------------------------------------------------

for layer in base_model.layers[140:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history_ft = model.fit(
    train_gen,
    epochs=3,
    validation_data=test_gen
)


# ---------------------------------------------------------
# 6. EVALUATE MODEL
# ---------------------------------------------------------

loss, acc = model.evaluate(test_gen)
print(f"\nFINAL TEST ACCURACY: {acc*100:.2f}%")


# ---------------------------------------------------------
# 7. GRAD-CAM HEATMAP (Explainability)
# ---------------------------------------------------------

def generate_gradcam(model, img_path, layer_name="conv5_block3_out"):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img_array = np.expand_dims(img/255.0, axis=0)

    grad_model = Model(
        inputs=model.input,
        outputs=[model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(img_array)
        class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]

    grads = tape.gradient(loss, conv_out)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))

    conv_out = conv_out[0]
    heatmap = conv_out @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = np.maximum(heatmap, 0) / np.max(heatmap)

    heatmap = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    superimposed = heatmap * 0.5 + img
    return img, superimposed.astype(np.uint8)


# EXAMPLE USAGE:
test_img = "/content/dataset/test/COVID/sample.png"

# Ensure the file exists for Grad-CAM example
if not os.path.exists(test_img):
    # Re-create dummy image if it was removed or not created initially
    create_dummy_dataset(img_size=IMG_SIZE) # Re-create the dataset

orig, cam = generate_gradcam(model, test_img)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(orig)
plt.axis(False)

plt.subplot(1,2,2)
plt.title("Grad-CAM Heatmap")
plt.imshow(cam)
plt.axis(False)

plt.show()


# ---------------------------------------------------------
# 8. SAVE MODEL FOR DEPLOYMENT
# ---------------------------------------------------------

model.save("covid_xray_resnet50.h5")
print("Model saved as covid_xray_resnet50.h5")
```
