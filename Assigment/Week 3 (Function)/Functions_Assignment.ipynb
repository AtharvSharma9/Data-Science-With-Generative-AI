**1. What is the difference between a function and a method in Python?**

-> In Python, both functions and methods are used to perform specific tasks, but they differ in how and where they are defined and used. A function is a block of reusable code that is defined independently using the def keyword and can be called by its name. It does not belong to any object unless explicitly passed. For example, a simple function like def greet(name): return "Hello, " + name can be called using greet("Atharv").

On the other hand, a method is a function that is associated with an object and is defined inside a class. It automatically takes the instance (self) as its first parameter and is called using the dot (.) operator. For instance, in a class Greeter, we might define a method def greet(self, name): return "Hello, " + name, and it would be called as greeter.greet("Atharv") after creating an object greeter = Greeter().


**Example of function**
```
def greet(name):
    return f"Hello, {name}!"

print(greet("Atharv"))

```
**Example of method**
```
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Atharv"))
```


**2.  Explain the concept of function arguments and parameters in Python.**

-> A parameter is a variable that is listed inside the parentheses in the function definition. It acts as a placeholder for the value that will be passed to the function. An argument, on the other hand, is the actual value that you provide to the function when calling it.
For example:
```
def greet(name):  # 'name' is a parameter
    print("Hello,", name)

greet("Atharv")    # "Atharv" is an argument
```
In the example above, name is the parameter, and "Atharv" is the argument passed during the function call.

Python also supports different types of arguments:

Positional Arguments – Values are passed in the same order as parameters.

Keyword Arguments – You specify which value goes to which parameter by name.

Default Arguments – Parameters can have default values if no argument is provided.

Variable-length Arguments – You can use *args for multiple positional arguments and **kwargs for multiple keyword arguments.

**5. What are iterators in Python and how do they differ from iterables? **

->
**What is an Iterable?**

An iterable is any Python object capable of returning its members one at a time. It implements the __iter__() method, which returns an iterator.

Examples of iterables:

* Lists ([1, 2, 3])

* Tuples ((1, 2, 3))

* Strings ("hello")

* Sets, Dictionaries, Ranges, etc.
```
my_list = [1, 2, 3]
for item in my_list:    # This works because lists are iterable
    print(item)
```

->
**What is an Iterator?**
An iterator is an object that represents a stream of data; it remembers the current position while iterating and returns the next value when next() is called.
It must implement:

* __iter__() (returns itself)

* __next__() (returns the next item)

```
my_list = [1, 2, 3]
my_iterator = iter(my_list)  # turns iterable into an iterator

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
```

**6. Explain the concept of generators in Python and how they are defined.**

->
**What Is a Generator?**
A generator is:

1. A function that uses yield instead of return

2. An iterator (it implements __iter__() and __next__())

3. Lazy-evaluated: it produces values only when requested

Defining a Generator Function:
```
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```
 Using the Generator:
 ```
 gen = count_up_to(3)

print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```


**7. What are the advantages of using generators over regular functions?**

->

1. Memory Efficiency

Generators yield one value at a time, so they don’t store the entire result in memory. This is especially useful when working with large datasets or infinite sequences.
```
def gen_nums():
    for i in range(1000000):
        yield i
```

2. Lazy Evaluation (On-demand computation)

Generators compute values only when needed, using the next() function. This reduces initial processing time and makes programs more efficient.

3. Can Represent Infinite Sequences

Since generators don’t compute all values at once, they are perfect for creating infinite data streams (like counters, live data feeds, etc.).
```
def infinite_counter():
    i = 0
    while True:
        yield i
        i += 1
```
4. Better Performance in Iterative Tasks

When only part of a result is needed (like looping until a condition is met), generators stop when done and avoid unnecessary processing.

5. Maintains Internal State Automatically

Generators remember their execution state, so when paused at yield, they can resume from the same point on the next call.

6. Cleaner, More Readable Code

Generators simplify code for streaming data, pipelines, and lazy evaluation scenarios. They are more readable than managing iteration manually.

**8. What is a lambda function in Python and when is it typically used?**

->
A lambda function in Python is a small, anonymous function defined with the lambda keyword instead of the usual def. It can take any number of arguments but can only have a single expression, which is evaluated and returned automatically.

Syntax: lambda arguments: expression

Example:
```
square = lambda x: x * x
print(square(5))  # Output: 25
```

**When is a lambda function typically used?**
* When you need a small, simple function for a short time without formally defining it.

* Often used as an inline function in places like:

    * map(), filter(), and sorted() functions.

    * As arguments to higher-order functions.

* When defining quick, throwaway functions for short operations.

**9. Explain the purpose and usage of the `map()` function in Python.**

->The map() function in Python is used to apply a given function to every item of an iterable (like a list, tuple, etc.) and return a map object (an iterator) of the results. It allows you to transform or process all elements in a sequence efficiently and concisely without writing an explicit loop.

Purpose of map()
* To apply the same function to all items in an iterable.

* To avoid writing loops for element-wise operations.

* To return a lazy iterator that produces results on demand, saving memory.

Syntax: map(function, iterable, ...)

Example:
```
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]
```
Using map() with Lambda:
```
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```
Using Multiple Iterables:
```
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

result = map(lambda x, y: x + y, numbers1, numbers2)
print(list(result))  # Output: [5, 7, 9]
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

->

map()

* Purpose: Applies a function to each item of one or more iterables and returns a map object (an iterator) with the transformed items.
* Input: A function and one or more iterables.
* Output: Iterator with transformed values.
* Use case: Transform or modify all elements.

Example:
```
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

filter()

* Purpose: Applies a function (that returns True or False) to each item of an iterable and returns an iterator with only the items where the function returns True.
* Input: A function that returns a boolean and one iterable.
* Output: Iterator with filtered items.
* Use case: Select items based on a condition.

Example:
```
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]
```
reduce()

* Purpose: Applies a rolling computation to sequential pairs of values in an iterable, reducing it to a single cumulative value.
* Input: A function and one iterable.
* Output: A single value.
* Use case: Aggregate or combine elements to a single result.
* Note: reduce() is in the functools module, so you need to import it:
from functools import reduce

Example:
```
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24 (1*2*3*4)
```

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
print(sum_of_evens([1, 2, 3, 4, 5, 6]))
````

````python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example usage:
print(square_numbers([1, 2, 3, 4]))
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage for numbers 1 to 200:
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.index = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.n_terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        else:
            self.a, self.b = self.b, self.a + self.b
            self.index += 1
            return self.a

# Example usage:
fib = Fibonacci(10)
print(list(fib))  # Output: First 10 Fibonacci numbers
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):
    print(power)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"File not found: {file_path}")

# Example usage:
for line in read_file_line_by_line('example.txt'):
    print(line)
````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples_list = [(1, 3), (3, 2), (5, 1)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_list = [0, 20, 37, 100]
fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
print(fahrenheit_list)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

# Example usage:
print(remove_vowels("This is a test."))
````
